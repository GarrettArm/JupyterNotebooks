```python
def clean_up_metadata(xml_text):
    for key, value in nickname_dict.items():
        xml_text = xml_text.replace('<{}>'.format(key), '<{}>'.format(value.replace(' ', '_').lower()))
        xml_text = xml_text.replace('</{}>'.format(key), '</{}>'.format(value.replace(' ', '_').lower()))
        xml_text = xml_text.replace('<{}/>'.format(key), '<{}/>'.format(value.replace(' ', '_').lower()))
    return xml_text
```

```python
# there's a much better way to do this using built-in's.  

import os

dir_class_parent = []
root_directory = '/home/james/Desktop/KivyProjects/venv_july/lib/python2.7/site-packages/'


def find_class_parents(directory):
    os.chdir(directory)
    for i in os.listdir():
        if i[-2:] == 'py':
            with open(i, 'r') as f:
                for line in f:
                    for word in line.split(' '):
                        if word == 'class':
                            if '):' in line:
                                namespace_etc = line[line.find('class') + 6:]
                                namespace = namespace_etc[:namespace_etc.find('(')]
                                class_name = namespace_etc[namespace_etc.find('('):]
                                for symb in (':', '(', ')', '\n'):
                                    class_name = class_name.replace(symb, '')
                                shortened_directory = os.getcwd().replace(root_directory, '')
                                dir_class_parent.append(
                                    [shortened_directory, i, namespace, class_name])


def check_if_directory(directory):
    os.chdir(directory)
    current_dir = os.getcwd()
    for i in os.listdir():
        if '.' not in i:
            done = False
            while not done:
                check_if_directory(current_dir + '/' + i)
                find_class_parents(os.getcwd())
                done = True


check_if_directory(root_directory + 'kivy/')

dir_class_parent = sorted(dir_class_parent, key = lambda x : (x[0], x[1], x[2], x[3]))
    
for i in dir_class_parent:
    print(i)
```
